```python
 for i in range(0, num_districts-1):
        r_vertex_1 = span_edges_nodes[cut_edges[i]][0]
        r_vertex_2 = span_edges_nodes[cut_edges[i]][1]
        temp[r_vertex_1][r_vertex_2] = 0
        temp[r_vertex_2][r_vertex_1] = 0
```
